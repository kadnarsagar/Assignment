1. What is hypothesis testing in statistics?

->>

Hypothesis testing is a fundamental concept in statistics that allows researchers to make inferences about a population based on sample data. It involves evaluating two competing hypotheses to determine which one is better supported by the data.

Hypothesis testing is a statistical method used to assess the validity of a claim or assumption about a population parameter. It involves formulating two mutually exclusive hypotheses:

* Null Hypothesis (H₀): Assumes no effect or difference exists in the population.
* Alternative Hypothesis (H₁ or Hₐ): Contradicts the null hypothesis, suggesting that an effect or difference does exist.

The goal is to use sample data to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis. This process helps researchers draw conclusions about populations without examining every individual.


2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

->>

In statistical hypothesis testing, the null hypothesis (H₀) and the alternative hypothesis (H₁ or Hₐ) are two mutually exclusive statements that form the foundation of inferential statistics.

Null Hypothesis (H₀)

The null hypothesis posits that there is no effect, no difference, or no relationship between the variables under study. It serves as the default assumption that any observed differences are due to chance or random variation.

Examples:
* In a clinical trial, H₀ might state that a new drug has no effect on patient recovery compared to a placebo.
* In a manufacturing process, H₀ could assert that the mean weight of products is equal to the target weight specified by the manufacturer.

Alternative Hypothesis (H₁ or Hₐ)

The alternative hypothesis contradicts the null hypothesis, suggesting that there is an effect, a difference, or a relationship present. It represents the researcher's claim or the hypothesis they aim to support through statistical evidence.

Examples:
* In the same clinical trial, H₁ might propose that the new drug improves patient recovery compared to a placebo.
* In the manufacturing scenario, H₁ could claim that the mean weight of products is not equal to the target weight.


3. What is the significance level in hypothesis testing, and why is it important?

->>

The significance level represents the probability of rejecting the null hypothesis when it is actually true. For instance, a significance level of 0.05 indicates a 5% risk of concluding that a difference exists when there is no actual difference. This means that, if the null hypothesis is true, there's a 5% chance of observing a result as extreme as, or more extreme than, the one obtained.

Why Is the Significance Level Important?
* Controls Type I Error Rate: By setting a significance level, researchers control the probability of making a Type I error, which occurs when a true null hypothesis is incorrectly rejected. This is crucial in ensuring the reliability of statistical conclusions.

* Establishes an Evidentiary Standard: The significance level acts as a predefined standard that the observed data must meet or exceed to be considered statistically significant. This helps in making consistent and objective decisions based on the data.

* Balances Risk of Errors: Choosing an appropriate significance level involves balancing the risk of Type I errors with the risk of Type II errors (failing to reject a false null hypothesis). In critical fields like medicine, a lower significance level (e.g., 0.01) may be chosen to minimize the risk of Type I errors, while in exploratory research, a higher significance level (e.g., 0.10) might be acceptable.


4. What does a P-value represent in hypothesis testing?

->>

In hypothesis testing, the p-value (probability value) quantifies the strength of evidence against the null hypothesis. It represents the probability of obtaining results at least as extreme as those observed, assuming the null hypothesis is true.

The p-value indicates how likely it is to observe the collected data, or something more extreme, under the assumption that the null hypothesis is correct. A smaller p-value suggests stronger evidence against the null hypothesis.

For example, if a clinical trial yields a p-value of 0.03, it means there is a 3% probability that the observed effect occurred by chance under the null hypothesis. Since this is less than the commonly used significance level of 0.05, the result is considered statistically significant, leading to the rejection of the null hypothesis.

6. What are Type 1 and Type 2 errors in hypothesis testing?

->>

Type I Error (False Positive)
* Definition: Occurs when the null hypothesis (H0) is true, but is incorrectly rejected.
* Example: A clinical trial concludes that a new drug is effective when, in reality, it is not.
* Probability: Denoted by α (alpha), commonly set at 0.05, indicating a 5% risk of committing this error.
* Implications: May lead to the adoption of ineffective treatments or interventions.

Type II Error (False Negative)
* Definition: Occurs when the null hypothesis is false, but fails to be rejected.
* Example: A clinical trial concludes that a new drug is not effective when, in reality, it is.
* Probability: Denoted by β (beta), with the power of the test being 1 - β.
* Implications: May result in the dismissal of beneficial treatments or interventions.


7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

->>

One-Tailed Test
* Definition: A one-tailed test evaluates the possibility of the relationship in only one direction—either greater than or less than a specified value.

* Alternative Hypothesis:
    * Right-Tailed (Upper-Tailed): Tests if a parameter is greater than a certain value.
    * Left-Tailed (Lower-Tailed): Tests if a parameter is less than a certain value.

* Critical Region: Located entirely in one tail of the distribution.
* Example: Testing if a new drug increases recovery rates.
    * Null Hypothesis (H₀): The drug does not increase recovery rates.
    * Alternative Hypothesis (H₁): The drug increases recovery rates.

Two-Tailed Test
* Definition: A two-tailed test evaluates the possibility of the relationship in both directions—whether a parameter is either greater than or less than a specified value.

* Alternative Hypothesis: Tests if a parameter is different from a certain value (either greater or less).

* Critical Regions: Divided between both tails of the distribution.

* Example: Testing if a new teaching method affects student performance (either improvement or decline).
    * Null Hypothesis (H₀): The teaching method has no effect.
    * Alternative Hypothesis (H₁): The teaching method affects performance (either positively or negatively).

10. What is the T-distribution, and when should it be used instead of the normal distribution?

->>

The t-distribution, also known as Student’s t-distribution, is a probability distribution used in statistics for hypothesis testing and confidence interval estimation, particularly when dealing with small sample sizes. It is similar to the normal distribution but has heavier tails, which means it accounts for more variability and extreme values.

Key Characteristics of the t-Distribution
* Symmetry: The t-distribution is symmetric around zero, just like the normal distribution.

* Heavier Tails: It has thicker tails than the normal distribution, providing more probability to extreme values. This feature becomes more pronounced with smaller sample sizes.

* Degrees of Freedom (df): The shape of the t-distribution depends on the degrees of freedom, typically calculated as the sample size minus one (n - 1). As the degrees of freedom increase, the t-distribution approaches the normal distribution.

* Mean and Variance: The mean of the t-distribution is zero. Its variance is greater than 1 and depends on the degrees of freedom.

t-Distribution
* Tail Behavior - Heavier tails
* Use Case - Small sample sizes, unknown population SD
* Convergence - Approaches normal as df → ∞
* Variance - Greater than 1, depends on df

Normal Distribution
* Tail Behavior - Lighter tails
* Use Case - Large sample sizes, known population SD
* Convergence - Fixed shape
* Variance - Fixed at 1

14. What is a confidence interval, and how is it used to interpret statistical results?

->>

A confidence interval (CI) is a statistical tool used to estimate the range within which a population parameter—such as a mean or proportion—is likely to fall, based on sample data. It provides a measure of uncertainty around the point estimate, offering more information than a single value alone.

A confidence interval is typically expressed as:

      Point Estimate ± Margin of Error

* Point Estimate: The sample statistic (e.g., sample mean) used as the best estimate of the population parameter.

* Margin of Error: A value that quantifies the uncertainty or variability in the sample estimate.

Interpreting Confidence Intervals

The confidence level, often set at 95%, indicates the probability that the interval contains the true population parameter if the sampling were repeated numerous times. It's important to note that this does not mean there's a 95% chance that the current interval contains the true parameter; rather, it reflects the long-term reliability of the estimation process.

17. What is the Chi-square distribution, and when is it used?

->>

The chi-square distribution is the distribution of a sum of the squares of independent standard normal variables. It is parameterized by the degrees of freedom (df), which typically correspond to the number of categories or groups minus one.

The distribution is right-skewed and becomes more symmetric as the degrees of freedom increase. The mean of the chi-square distribution is equal to its degrees of freedom, and its variance is twice the degrees of freedom.

The chi-square distribution is primarily used in the following statistical tests:

1. Chi-Square Goodness of Fit Test:
  * Purpose: To determine if the observed frequencies of a categorical variable match the expected frequencies based on a specific hypothesis.
  * Example: Testing whether a die is fair by comparing the observed frequency of each face to the expected frequency.

2. Chi-Square Test of Independence:
  * Purpose: To assess whether two categorical variables are independent or associated.
  * Example: Examining if gender and voting preference are independent in a population.

3. Chi-Square Test for Homogeneity:
  * Purpose: To compare the distribution of a categorical variable across different populations or groups.
  * Example: Comparing the distribution of blood types across different regions.

19. What is the F-distribution, and when is it used in hypothesis testing?

->>

Definition: The F-distribution is the distribution of a ratio of two scaled chi-square distributions. Specifically, if you have two independent chi-square variables, each divided by their respective degrees of freedom, the ratio follows an F-distribution.

1. ANOVA (Analysis of Variance):
  * Purpose: To compare the means of three or more groups to determine if at least one group mean is significantly different from the others.
  * Test Statistic: The F-statistic is calculated as the ratio of the variance between group means to the variance within groups.
  * Hypotheses:
      * Null Hypothesis (H₀): All group means are equal.
      * Alternative Hypothesis (H₁): At least one group mean is different.
  * Decision Rule: If the calculated F-statistic exceeds the critical value from the F-distribution table (based on df₁ and df₂), reject the null hypothesis.

2. Regression Analysis:
  * Purpose: To assess the overall significance of a regression model.
  * Test Statistic: The F-statistic tests whether at least one predictor variable has a non-zero coefficient.
  * Hypotheses:
      * Null Hypothesis (H₀): All regression coefficients are zero (no relationship between predictors and the dependent variable).
      * Alternative Hypothesis (H₁): At least one regression coefficient is non-zero.
  * Decision Rule: A significant F-statistic indicates that the model explains a significant portion of the variance in the dependent variable.

3. Equality of Two Variances:
  * Purpose: To compare the variances of two populations.
  * Test Statistic: The ratio of the two sample variances follows an F-distribution.
  * Hypotheses:
      * Null Hypothesis (H₀): The population variances are equal.
      * Alternative Hypothesis (H₁): The population variances are not equal.
  * Decision Rule: If the calculated F-statistic falls in the critical region of the F-distribution, reject the null hypothesis.



20. What is an ANOVA test, and what are its assumptions?

->>

Analysis of Variance (ANOVA) is a statistical method used to compare means across multiple groups to determine if at least one group mean is significantly different from the others. It assesses the impact of one or more independent variables on a dependent variable by analyzing the variance within and between groups.

Key Assumptions of ANOVA

To ensure the validity of ANOVA results, certain assumptions must be met:

  1. Normality: The data within each group should follow a normal distribution. This assumption is particularly important when sample sizes are small.

  2. Independence: Observations within each group must be independent of each other. This means the value of one observation should not influence another.

  3. Homogeneity of Variances (Homoscedasticity): The variance within each group should be approximately equal. This can be tested using Levene's test or by comparing the standard deviations across groups.

21. What are the different types of ANOVA tests?

->>

1. One-Way ANOVA
  * Purpose: Compares means across three or more independent groups based on a single factor.
  * Example: Assessing the effect of different diets on weight loss.
  * Assumptions:
    * Independence of observations.
    * Normality of data within each group.
    * Homogeneity of variances (equal variances across groups)
2. Two-Way ANOVA
  * Purpose: Examines the influence of two independent variables on a dependent variable, and also assesses the interaction between these variables.
  * Example: Studying the effects of exercise type and diet on weight loss.
  * Assumptions:
    * Independence of observations.
    * Normality of data within each group.
    * Homogeneity of variances.
    * Additivity and linearity of effects.
3. Factorial ANOVA
  * Purpose: Extends two-way ANOVA to include more than two independent variables, allowing for the examination of interactions among multiple factors.
  * Example: Investigating the combined effects of exercise type, diet, and sleep duration on weight loss.
  * Assumptions:
    * Independence of observations.
    * Normality of data within each group.
    * Homogeneity of variances.
    * Additivity and linearity of effects.

22. What is the F-test, and how does it relate to hypothesis testing?

->>

An F-test evaluates the ratio of variances derived from two different sources:

* Numerator: Variance between group means (or explained variance).
* Denominator: Variance within groups (or unexplained variance).

The F-statistic is calculated as:

    F = variance within groups / Variance between groups

Under the null hypothesis (H₀), which posits that all group means are equal, the F-statistic follows an F-distribution. A significantly large F-statistic suggests that at least one group mean differs from the others.

* * F-Test in Hypothesis Testing

In the context of hypothesis testing, the F-test is used to:

  * Compare Multiple Group Means: In Analysis of Variance (ANOVA), the F-test assesses whether there are any statistically significant differences between the means of three or more independent groups.

  * Evaluate Regression Models: In multiple regression analysis, the F-test determines whether at least one predictor variable has a non-zero coefficient, indicating that the model explains a significant portion of the variance in the dependent variable.

The general hypotheses for an F-test are:

  * Null Hypothesis (H₀): All group means are equal (or all regression coefficients are zero).

  * Alternative Hypothesis (H₁): At least one group mean is different (or at least one regression coefficient is non-zero).

A small p-value (typically < 0.05) leads to rejecting the null hypothesis, suggesting that the model or group differences are statistically significant.

2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
low, high = 1, 6

# Generate 1000 random samples
samples = np.random.randint(low, high + 1, size=1000)

# Calculate the PMF
values, counts = np.unique(samples, return_counts=True)
probabilities = counts / len(samples)

# Plot the PMF
plt.bar(values, probabilities, width=0.8, align='center', alpha=0.7)
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Mass Function of Discrete Uniform Distribution')
plt.xticks(values)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pmf(x, p):
    """
    Calculate the PMF of a Bernoulli distribution.

    Parameters:
    x : int (should be 0 or 1)
        The outcome for which to calculate the probability.
    p : float
        The probability of success (1), where 0 <= p <= 1.

    Returns:
    float : The probability of observing outcome x.
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p if x == 1 else 1 - p
```

```python
# Example: probability of success is 0.7
print("P(X=1):", bernoulli_pmf(1, 0.7))  # Output: 0.7
print("P(X=0):", bernoulli_pmf(0, 0.7))  # Output: 0.3
```

9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculate Z-scores for the data and plot them.

    Parameters:
    data (array-like): The dataset (list, numpy array, etc.)

    Returns:
    z_scores (np.ndarray): Array of Z-scores.
    """
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data)

    if std_dev == 0:
        raise ValueError("Standard deviation is zero; Z-scores cannot be calculated.")

    z_scores = (data - mean) / std_dev

    # Plot Z-scores
    plt.figure(figsize=(8, 5))
    plt.scatter(range(len(z_scores)), z_scores, color='blue', alpha=0.7)
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.axhline(1, color='green', linestyle='--', label='Z=1')
    plt.axhline(-1, color='green', linestyle='--', label='Z=-1')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.title('Z-scores of the Dataset')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage
data = [10, 12, 23, 23, 16, 23, 21, 16]
zscores = calculate_and_plot_zscores(data)
print("Z-scores:", zscores)
```

15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, mu0, sigma, alpha=0.05):
    """
    Perform a two-tailed Z-test for a sample dataset.

    Parameters:
    - sample: array-like, sample data
    - mu0: hypothesized population mean
    - sigma: known population standard deviation
    - alpha: significance level (default 0.05)

    Returns:
    - z_stat: calculated Z statistic
    - p_value: two-tailed p-value
    - reject_null: True if null hypothesis is rejected, else False
    """
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate Z statistic
    z_stat = (sample_mean - mu0) / (sigma / np.sqrt(n))

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Determine whether to reject the null hypothesis
    reject_null = p_value < alpha

    return z_stat, p_value, reject_null

# Example usage
sample_data = [102, 98, 100, 105, 110, 95, 99, 101, 104, 97]
hypothesized_mean = 100
known_sigma = 5
significance_level = 0.05

z_stat, p_val, reject = z_test(sample_data, hypothesized_mean, known_sigma, significance_level)

print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.4f}")
print(f"Reject Null Hypothesis? {'Yes' if reject else 'No'}")
```

16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of the dataset.

    Parameters:
    - data: array-like sample data
    - confidence: confidence level (default 0.95)

    Returns:
    - (lower_bound, upper_bound): confidence interval limits
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # Standard error of the mean

    # t critical value for two-tailed test
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * sem
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example data
sample_data = [102, 98, 100, 105, 110, 95, 99, 101, 104, 97]

# Calculate 95% CI
lower, upper = confidence_interval(sample_data, confidence=0.95)

print(f"Sample mean = {np.mean(sample_data):.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
print("\nInterpretation:")
print(f"We are 95% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")
```

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Perform a two-tailed Z-test comparing sample mean to population mean.

    Parameters:
    - sample: array-like sample data
    - population_mean: known population mean (H0 mean)
    - population_std: known population standard deviation
    - alpha: significance level (default 0.05)

    Returns:
    - z_stat: calculated Z statistic
    - p_value: two-tailed p-value
    - result: interpretation string
    """
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate Z statistic
    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Interpretation
    if p_value < alpha:
        result = f"Reject the null hypothesis at α={alpha}. Sample mean ({sample_mean:.2f}) "\
                 f"is significantly different from population mean ({population_mean})."
    else:
        result = f"Fail to reject the null hypothesis at α={alpha}. Sample mean ({sample_mean:.2f}) "\
                 f"is not significantly different from population mean ({population_mean})."

    return z_stat, p_value, result

# Example usage
sample_data = [102, 98, 100, 105, 110, 95, 99, 101, 104, 97]
pop_mean = 100
pop_std = 5
alpha_level = 0.05

z_stat, p_val, interpretation = z_test(sample_data, pop_mean, pop_std, alpha_level)

print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.4f}")
print(interpretation)
```

3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

```python
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform a one-sample Z-test.

    Parameters:
    - sample: array-like sample data
    - pop_mean: population mean to compare against
    - pop_std: population standard deviation (known)
    - alpha: significance level (default 0.05)

    Returns:
    - z_stat: calculated Z statistic
    - p_value: two-tailed p-value
    - conclusion: interpretation string
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    standard_error = pop_std / np.sqrt(n)

    z_stat = (sample_mean - pop_mean) / standard_error
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # two-tailed test

    if p_value < alpha:
        conclusion = (f"Reject the null hypothesis at α={alpha}. "
                      f"Sample mean ({sample_mean:.2f}) is significantly different from population mean ({pop_mean}).")
    else:
        conclusion = (f"Fail to reject the null hypothesis at α={alpha}. "
                      f"Sample mean ({sample_mean:.2f}) is not significantly different from population mean ({pop_mean}).")

    return z_stat, p_value, conclusion

# Example usage
sample_data = [105, 98, 100, 102, 97, 101, 99, 103, 104, 96]
population_mean = 100
population_std = 5  # known population standard deviation
alpha_level = 0.05

z_stat, p_val, result = one_sample_z_test(sample_data, population_mean, population_std, alpha_level)

print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.4f}")
print(result)
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    se = pop_std / np.sqrt(n)

    # Calculate Z statistic
    z_stat = (sample_mean - pop_mean) / se

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Critical z-values for two-tailed test
    z_crit = norm.ppf(1 - alpha/2)

    # Interpretation
    if abs(z_stat) > z_crit:
        conclusion = f"Reject null hypothesis at α={alpha}."
    else:
        conclusion = f"Fail to reject null hypothesis at α={alpha}."

    return z_stat, p_value, z_crit, conclusion

# Example data
sample_data = [104, 99, 101, 103, 98, 100, 97, 102, 105, 96]
population_mean = 100
population_std = 4
alpha_level = 0.05

z_stat, p_val, z_crit, conclusion = two_tailed_z_test(sample_data, population_mean, population_std, alpha_level)

print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.4f}")
print(f"Critical Z-value ±{z_crit:.3f}")
print(conclusion)

# Plotting the standard normal distribution with rejection regions
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution')

# Shade rejection regions
plt.fill_between(x, 0, y, where=(x <= -z_crit), color='red', alpha=0.3, label='Rejection Region (left)')
plt.fill_between(x, 0, y, where=(x >= z_crit), color='red', alpha=0.3, label='Rejection Region (right)')

# Mark the test statistic
plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z-statistic = {z_stat:.3f}')

plt.title('Two-tailed Z-test Decision Regions')
plt.xlabel('Z value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(mu0, mu1, sigma, n, alpha=0.05):
    """
    Calculate and visualize Type I and Type II errors.

    Parameters:
    - mu0: mean under null hypothesis
    - mu1: mean under alternative hypothesis
    - sigma: population standard deviation (known)
    - n: sample size
    - alpha: significance level (default 0.05)
    """
    se = sigma / np.sqrt(n)  # standard error

    # Critical value (two-tailed)
    z_crit = norm.ppf(1 - alpha/2)
    lower_crit = mu0 - z_crit * se
    upper_crit = mu0 + z_crit * se

    # Type I error (alpha) is fixed by significance level

    # Type II error calculation
    # Probability that test statistic falls within acceptance region given alternative mean mu1
    beta = norm.cdf(upper_crit, loc=mu1, scale=se) - norm.cdf(lower_crit, loc=mu1, scale=se)
    power = 1 - beta

    # Plotting distributions
    x_min = min(mu0, mu1) - 4*se
    x_max = max(mu0, mu1) + 4*se
    x = np.linspace(x_min, x_max, 1000)

    null_pdf = norm.pdf(x, mu0, se)
    alt_pdf = norm.pdf(x, mu1, se)

    plt.figure(figsize=(10,6))

    # Null hypothesis distribution
    plt.plot(x, null_pdf, label='Null Hypothesis (H0)', color='blue')
    # Alternative hypothesis distribution
    plt.plot(x, alt_pdf, label='Alternative Hypothesis (H1)', color='green')

    # Shade rejection regions for Type I error (two tails)
    plt.fill_between(x, 0, null_pdf, where=(x <= lower_crit), color='red', alpha=0.3, label='Type I Error Region (α)')
    plt.fill_between(x, 0, null_pdf, where=(x >= upper_crit), color='red', alpha=0.3)

    # Shade Type II error region under alternative hypothesis curve (acceptance region)
    plt.fill_between(x, 0, alt_pdf, where=((x > lower_crit) & (x < upper_crit)), color='orange', alpha=0.3, label='Type II Error Region (β)')

    # Critical values lines
    plt.axvline(lower_crit, color='red', linestyle='--')
    plt.axvline(upper_crit, color='red', linestyle='--')

    plt.title(f'Type I and Type II Errors\nα = {alpha}, β = {beta:.4f}, Power = {power:.4f}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Significance level (Type I error, α): {alpha}")
    print(f"Type II error (β): {beta:.4f}")
    print(f"Statistical Power (1 - β): {power:.4f}")

# Example usage:
plot_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05)
```

6. Write a Python program to perform an independent T-test and interpret the results.

```python
import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(sample1, sample2, alpha=0.05):
    """
    Perform an independent two-sample t-test and interpret the result.

    Parameters:
    - sample1: array-like data for group 1
    - sample2: array-like data for group 2
    - alpha: significance level (default 0.05)

    Returns:
    - t_stat: calculated t statistic
    - p_value: p-value of the test
    - interpretation: string interpreting the result
    """
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=False)  # Welch's t-test

    if p_value < alpha:
        interpretation = (f"Reject the null hypothesis at α={alpha}. "
                          f"There is a significant difference between the means "
                          f"of the two groups (p = {p_value:.4f}).")
    else:
        interpretation = (f"Fail to reject the null hypothesis at α={alpha}. "
                          f"No significant difference between the means "
                          f"of the two groups (p = {p_value:.4f}).")

    return t_stat, p_value, interpretation

# Example usage
np.random.seed(0)
group1 = np.random.normal(loc=20, scale=5, size=30)
group2 = np.random.normal(loc=22, scale=5, size=30)

t_stat, p_val, result = independent_t_test(group1, group2)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(result)
```

8. Simulate data and perform both Z-test and T-test, then compare the results using Python.

```python
import numpy as np
from scipy.stats import norm, ttest_1samp

# Set seed for reproducibility
np.random.seed(123)

# Parameters
pop_mean = 50
pop_std = 8          # Known population std dev for Z-test
sample_size = 30
alpha = 0.05

# Simulate sample data (unknown population std dev to t-test)
sample = np.random.normal(loc=52, scale=pop_std, size=sample_size)

# --- One-sample Z-test ---
sample_mean = np.mean(sample)
se_z = pop_std / np.sqrt(sample_size)
z_stat = (sample_mean - pop_mean) / se_z
p_value_z = 2 * (1 - norm.cdf(abs(z_stat)))

# --- One-sample T-test ---
t_stat, p_value_t = ttest_1samp(sample, pop_mean)

# Print results
print("One-sample Z-test:")
print(f"  Sample mean = {sample_mean:.3f}")
print(f"  Z-statistic = {z_stat:.3f}")
print(f"  P-value = {p_value_z:.4f}\n")

print("One-sample T-test:")
print(f"  Sample mean = {sample_mean:.3f}")
print(f"  T-statistic = {t_stat:.3f}")
print(f"  P-value = {p_value_t:.4f}\n")

# Interpretation
def interpret_test(p_value, test_name):
    if p_value < alpha:
        print(f"{test_name}: Reject the null hypothesis (significant difference).")
    else:
        print(f"{test_name}: Fail to reject the null hypothesis (no significant difference).")

interpret_test(p_value_z, "Z-test")
interpret_test(p_value_t, "T-test")
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    """
    Calculate the confidence interval for the sample mean.

    Parameters:
    - sample: array-like sample data
    - confidence: confidence level (default 0.95 for 95% CI)

    Returns:
    - (lower_bound, upper_bound): tuple of confidence interval bounds
    """
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # sample standard error

    # t critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
sample_data = [12, 15, 14, 16, 13, 15, 14, 17, 16, 14]
lower, upper = confidence_interval(sample_data, confidence=0.95)

print(f"Sample mean: {np.mean(sample_data):.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for a sample mean at a given confidence level.

    Parameters:
    - sample: array-like sample data
    - confidence: confidence level (default 0.95)

    Returns:
    - margin_of_error: the margin of error for the confidence interval
    """
    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)  # standard error of the mean
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)  # t critical value

    return t_crit * std_err

# Example usage
sample_data = [12, 15, 14, 16, 13, 15, 14, 17, 16, 14]
confidence_level = 0.95

moe = margin_of_error(sample_data, confidence_level)
print(f"Margin of Error at {confidence_level*100}% confidence: {moe:.3f}")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior, likelihood, false_positive_rate):
    """
    Calculate posterior probability using Bayes' Theorem.

    Parameters:
    - prior: P(H) prior probability of hypothesis (e.g., disease prevalence)
    - likelihood: P(E|H) probability of evidence given hypothesis (e.g., sensitivity)
    - false_positive_rate: P(E|¬H) probability of evidence given no hypothesis (1 - specificity)

    Returns:
    - posterior: P(H|E) posterior probability given evidence
    """
    # Total probability of evidence P(E)
    prob_evidence = likelihood * prior + false_positive_rate * (1 - prior)

    # Bayes' Theorem
    posterior = (likelihood * prior) / prob_evidence

    return posterior

# Given data
prior_disease = 0.01            # 1% prevalence
sensitivity = 0.99              # P(positive test | disease)
specificity = 0.95              # P(negative test | no disease)
false_positive_rate = 1 - specificity

posterior_prob = bayes_theorem(prior_disease, sensitivity, false_positive_rate)

print(f"Posterior probability of disease given positive test: {posterior_prob:.4f} ({posterior_prob*100:.2f}%)")
```

12. Perform a Chi-square test for independence between two categorical variables in Python.

```python
import numpy as np
from scipy.stats import chi2_contingency

# Example contingency table: rows = Variable A categories, columns = Variable B categories
# For instance: survey results of two groups and their preferences
observed = np.array([[30, 10, 20],
                     [20, 20, 15]])

# Perform Chi-square test for independence
chi2_stat, p_value, dof, expected = chi2_contingency(observed)

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p_value:.4f}")
print("Expected Frequencies:")
print(expected)

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: Variables are dependent (associated).")
else:
    print("Fail to reject null hypothesis: Variables are independent.")
```

13.  Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data.

```python
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test of independence.

    Parameters:
    - observed: 2D numpy array of observed counts

    Returns:
    - expected: 2D numpy array of expected counts
    """
    # Calculate row totals and column totals
    row_totals = observed.sum(axis=1).reshape(-1, 1)
    col_totals = observed.sum(axis=0).reshape(1, -1)
    total = observed.sum()

    # Expected frequency = (row total * column total) / grand total
    expected = (row_totals @ col_totals) / total
    return expected

# Example observed data
observed = np.array([[30, 10, 20],
                     [20, 20, 15]])

expected = calculate_expected_frequencies(observed)

print("Observed Frequencies:")
print(observed)
print("\nExpected Frequencies:")
print(expected)
```

14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import numpy as np
from scipy.stats import chisquare

# Observed data (e.g., counts in categories A, B, C, D)
observed = np.array([50, 30, 15, 5])

# Expected probabilities for each category (must sum to 1)
expected_probs = np.array([0.4, 0.3, 0.2, 0.1])

# Convert expected probabilities to expected counts (same total as observed)
expected_counts = expected_probs * observed.sum()

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected_counts)

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: observed data does NOT fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: observed data fits the expected distribution.")
```
