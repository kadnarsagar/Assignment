4. Discuss the different types of operators in Python and provide examples of how they are used.

->>  
         There are 7 types of operators used in python.

      1- Arithmatic operators                   Examples
            - Addition (+)             # 4 + 8   = 12
            - Subtraction (-)          # 6 - 23  = 17
            - Division (/)             # 10 / 2  = 5.0
            - Multiplication (*)       # 3 * 7   = 21
            - Modulus (%)              # 10 % 3  = 1
            - Exponentiation (**)      # 2 ** 3  = 8
            - Floor division (//)      # 10 // 3 = 3

      2- Comparison Operators

            - Equal to (==)          # 9 == 9    -> True
            - Not equal to (!=)      # 4 != 5    -> True
            - Greater than (>)       # 7 > 9     -> False
            - Less than (<)          # 12 < 56   -> True
            - Greater than equal to (>=)   # 34 >= 18  -> True
            - Less than equal to (<=)    # 12 <= 65  -> True

      3- Logical Operators

            - Logical AND     # 4 < 5 and 1 > 2  -> False
            - Logical OR      # 4 < 5 or 7 > 10  -> True
            - Logical NOT    # not(7 < 5)   -> True

      4- Bitwise operators

            - Bitwise AND (&)     # 18 & 3    -> 2
            - Bitwise OR (|)      # 3 | 5     -> 7
            - Bitwise Not         # ~ 3       -> -4
            - Bitwise XOR (^)     # 5 ^ 3     -> 6
            - Left shift (<<)     # 35 << 3   -> 280
            - Right shift (>>)    # 10 >> 1   -> 5

      5- Assignment operators

            - Assignment (=)                    # a = 10
            - Addition Assignment (+=)          # 5 += 8    -> 13
            - Subtraction Assignment (-=)       # 12 -= 8   -> 4
            - Multiplication Assignment (*=)    # 4 *= 5    -> 20
            - Division Assignment (/=)          # 12 /= 6   -> 2

      6- Membership operators

            - in       # a in apple      -> True
            - not in   # a not in apple  -> False

      7- Identity Operators

            - is       # 4 is 4  - > True
            - is not   # 5 not 4 - > True

5. Explain the concept of type casting in Python with examples.

->>  
 Type Casting or Type Conversion is the process of changing the data ntype of a value / object to another data type.

      1. Implicit type casting / conversion.

            Python understand the data type without user intervention. (Automatically done).

            Example -
                  a = 5
                  type(a)       # int
                  b = 2 + 3.6   # 5.6
                  type(b)       # float

      2. Explicit type casting / conversioin.
      
            User will be used function (Manually done) int, float, ste, etc.

            Example -

                a = 9
                type(a)    # int
                float(a)   # 9.0
                type(a)    # float
                str(a)     # '9'
                type(a)    # str
